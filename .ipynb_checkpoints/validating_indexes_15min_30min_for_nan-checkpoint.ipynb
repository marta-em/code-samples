```python
from multiprocessing import Pool

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from scipy import spatial
from datetime import timedelta
import seaborn as sns

import sys
import collections
import itertools
from scipy.stats import mode

from numpy import *
from pylab import *
from scipy.spatial import distance
from sklearn import metrics


df = pd.read_csv('D:\data from MnM Samdal raw\Samdal_power_15min_no_outliers.csv', parse_dates=[0])

df.power = df.power.interpolate(method='linear', limit = 2 )
df.speed = df.speed.interpolate(method='linear', limit = 2 )

df = df.rename( columns={"Unnamed: 0": "time"})

# discard April
from datetime import timedelta
df = df.drop(df[df.time.dt.month==4].index)

df.time = pd.to_datetime(df.time, format='%Y%m%d %H:%M:%s')
df = df.set_index('time')

df = df.dropna(subset=['speed'])

############

# take 20% after dropping nan

# for 1h = 580 hrs (2902 total)  => 580/24 I need 24 indexes
# for 30min = 1151 30min = 575 hrs => 1151/24 (12 h ahead) I need 47 indexes
# for 15min = 2286 15min = 571 hrs => 2286 (6hrs aead) 95 indexes
```

```python
# finding bad indexes with nan in a subseries

wrong_idx = []

for i in df_indexes:
    subseries = df.speed[i:i+24] 
    first_idx = subser.index[0]
    last_idx = subser.index[-1]

    diff = last_idx - first_idx 
    tdt = timedelta(days=0, hours =23)  # correct case
    
    if diff != tdt:
        wrong_idx.append(i)
    
wrong_idx
```
